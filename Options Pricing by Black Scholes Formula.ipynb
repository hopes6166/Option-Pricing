```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
%matplotlib inline
```

```python
#draw Option Price graph based on Strike Price
#K is strike price
K = np.arange(.1, 20, .05)
#r is risk free rate
r = 0.07
#T is time to expiry
T = 1
#delta is compound dividend yield
delta = 0
#S0 is current stock price
S0 = 10
#sigma is annual volatility
sigma = .5
def Call(S0, K, T, r, delta, sigma):
    d1 = (np.log(S0/K) + (r- delta + 0.5*sigma**2)*T)/(sigma*np.sqrt(T))
    d2 = d1 -sigma*np.sqrt(T)
    return S0*np.exp(-delta*T)*norm.cdf(d1) - K*np.exp(-r*T)*norm.cdf(d2)

def Put(S0, K, T, r, delta, sigma):
    d1 = (np.log(S0/K) + (r- delta + 0.5*sigma**2)*T)/(sigma*np.sqrt(T))
    d2 = d1 -sigma*np.sqrt(T)
    return K*np.exp(-r*T)*norm.cdf(-d2) - S0*np.exp(-delta*T)*norm.cdf(-d1)

C = Call(S0, K, T, r, delta, sigma)
P = Put(S0, K, T, r, delta, sigma)
plt.plot(K,C,'r', K, P, 'g')
plt.axis([0,22, 0,11])
plt.title("Strike price and Options pricing based on Black Scholes Formula")
plt.xlabel("Strike price")
plt.ylabel("Option price")
plt.legend(["Call option", "Put option"])
```
