```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
%matplotlib inline
```

```python
K = np.arange(.1, 20, .05)
r = 0.07
T = 1
delta = 0
S0 = 10
sigma = .5
d1 = (np.log(S0/K) + (r- delta + 0.5*sigma**2)*T)/(sigma*np.sqrt(T))
d2 = d1 - sigma*np.sqrt(T)
C = S0*np.exp(-delta*T)*norm.cdf(d1) - K*np.exp(-r*T)*norm.cdf(d2)
P = K*np.exp(-r*T)*norm.cdf(-d2) - S0*np.exp(-delta*T)*norm.cdf(-d1)
plt.plot(K,C,'r', K, P, 'g')
plt.axis([0,22, 0,11])
plt.title("Strike price and Options pricing based on Black Scholes Formula")
plt.xlabel("Strike price")
plt.ylabel("Option price")
```
